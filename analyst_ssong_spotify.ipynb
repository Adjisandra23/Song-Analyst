```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Spotify_Songs_2024.csv')
```

```python
df
```

```python
df= df.drop_duplicates(subset=['Track','Artist'])
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'])
```

```python
df.info()
```

```python
df['Year'] = df['Release Date'].dt.strftime('%Y')
df.groupby('Year')['Track'].size().tail(10)
```

```python
release_year = df.groupby('Year')['Track'].size().tail(10)
index_release = release_year.index
release_year.plot(kind='bar', color='#6787f1')
release_year.plot(kind='line', marker='>', color='#0f8a0c')
plt.xticks(ticks=range(10), labels=index_release)
plt.title('Total Release per Year in Spotify')
plt.ylabel('(Total)', fontweight='bold')
plt.xlabel('(Year)',fontweight='bold')
plt.show()
```

## Mencari dan membuat grafik berdasarkan score tertinggi

```python
df[['Track','Track Score']].sort_values(by='Track Score',ascending=False).head(10)
```

```python
higest_score = df[['Track','Track Score']].sort_values(by='Track Score',ascending=False).head(10)
warna_terang = ['#FFFF00', # Kuning
                '#FF7F50', # Merah muda cerah
                '#00FF7F', # Hijau muda
                '#00FFFF', # Cyan
                '#FF00FF', # Magenta
                '#f5d43d', # Kuning muda
                '#7FFF00', # Hijau chartreuse
                '#FF007F', # Merah muda fuchsia
                '#00FFFF', # Cyan (lagi, untuk variasi)
                '#FFD700'] # Emas
higest_score.plot(kind='barh', x='Track', y='Track Score', color=warna_terang)
plt.title('Track Score')
plt.legend('')
plt.ylabel('')
plt.gca().invert_yaxis()
```

### The most stream in Spotify

```python
df_copy = df.copy()
df = df_copy.drop_duplicates()
df_copy['Spotify Streams'].dropna(inplace=True)
df_copy
```

```python
df_copy['Spotify Streams'] = df['Spotify Streams'].str.replace(',','')
df_copy['Spotify Streams'] = df_copy['Spotify Streams'].fillna(value=0)
```

```python
df_copy['Spotify Streams'] = df_copy['Spotify Streams'].astype('int64')
```

```python
spotify_stream = df_copy[['Artist','Track','Spotify Streams']].sort_values(by='Spotify Streams',ascending=False).head(10)
spotify_stream = spotify_stream.reset_index(drop=True)
spotify_stream
```

```python
spotify_stream = df_copy[['Artist','Track','Spotify Streams']].sort_values(by='Spotify Streams',ascending=False).head(10)
spotify_stream = spotify_stream.reset_index(drop=True)
spotify_stream
```

```python
df_copy['combine'] = df_copy['Artist'] + ' ('+df['Track'] + ')'
df_copy[['combine','Spotify Streams']].sort_values(by='Spotify Streams', ascending=False).set_index('combine').head(10)
```

```python
combine = df_copy[['combine','Spotify Streams']].sort_values(by='Spotify Streams', ascending=False).set_index('combine').head(10)
combine.plot(kind='barh', color='#05be28')
plt.legend('')
plt.gca().invert_yaxis()
plt.ylabel('')
plt.title('TOP Spotify Streams', fontweight='bold')
```

### Spotify Playlist Count

```python
df_copy['Spotify Playlist Count'] = df['Spotify Playlist Count'].str.replace(',','')

df_copy['Spotify Playlist Count'] = df_copy['Spotify Playlist Count'].fillna(0)

df_copy['Spotify Playlist Count'] = df_copy['Spotify Playlist Count'].astype('int64')

df_copy[['Track','Spotify Playlist Count']].sort_values(by='Spotify Playlist Count',ascending=False).head(10).set_index('Track')
```

```python
playlist_count = df_copy[['Track','Spotify Playlist Count']].sort_values(by='Spotify Playlist Count',ascending=False).head(10)
playlist_count.plot(kind='barh', x='Track',y='Spotify Playlist Count')
plt.gca().invert_yaxis()
plt.ylabel('')
plt.legend('')
plt.title('TOP Spotify Playlist', fontweight='bold')
```

```python
df.groupby('Artist')['Track'].count().sort_values(ascending=False).head(10)
```

```python
artist_song = df.groupby('Artist')['Track'].count().sort_values(ascending=False).head(10)
colors = [
    '#1f77b4',  # Biru
    '#ff7f0e',  # Oranye
    '#2ca02c',  # Hijau
    '#d62728',  # Merah
    '#9467bd',  # Ungu
    '#8c564b',  # Coklat
    '#e377c2',  # Pink
    '#7f7f7f',  # Abu-abu
    '#bcbd22',  # Kuning Zaitun
    '#17becf'   # Biru Muda
]
artist_song.plot(kind='bar',color=colors)
plt.xticks(rotation=45, ha='right')
plt.title('Artist dengan Lagu terbanyak di Spotify', fontweight='bold')
plt.show()
```
